```python
import requests
import bs4 as bs
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
userName = input("Enter valid github user name")

HEADERS = ({'User-Agent':
			'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/44.0.2403.157 Safari/537.36',
			'Accept-Language': 'en-US, en;q=0.5'})

following = "https://github.com/"+str(userName)+"?tab=following"

follower = "https://github.com/"+str(userName)+"?tab=followers"

follower_list = []
following_list = []

follower_set = set()
following_set = set()

# HTTP Request
following_request = requests.get(following, headers=HEADERS)
follower_request = requests.get(follower, headers=HEADERS)

# Soup Object containing all data
following_page = bs(following_request.content, "lxml")
follower_page = bs(follower_request.content, "lxml")

following_name = following_page.find_all("span",{"class": "f4 Link--primary"})
following_username = following_page.find_all("span",{"class": "Link--secondary"})

print("People who you are following\n")

for (name,username) in zip(following_name,following_username):
    following_list.append([name.string,username.string])
    follower_set.add(username.string)

following_df = pd.DataFrame(following_list, columns=['Following Name', 'Following Username'])
display(following_df)

follower_name = follower_page.find_all("span",{"class": "f4 Link--primary"})
follower_username = follower_page.find_all("span",{"class": "Link--secondary"})

print("\n\nPeople who follow you\n")

for (name,username) in zip(follower_name,follower_username):
	follower_list.append([name.string,username.string])
	following_set.add(username.string)

follower_df = pd.DataFrame(follower_list, columns=['Name', 'username'])
display(follower_df)

print("\nPeople the user is not following : \n","\n".join(list(following_set - follower_set)))
print("\nPeople who are not following the user : \n","\n".join(list(follower_set - following_set)))
print("\nBoth following and follower\n","\n".join(list(follower_set.intersection(following_set))))
```
